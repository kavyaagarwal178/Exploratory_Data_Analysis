**1- Importing the required libraries for EDA**

```python
import pandas as pd
import numpy as np
import seaborn as sns                      
import matplotlib.pyplot as plt 
```

**2- Loading the data into data frame**

```python
df = pd.read_csv("/content/Google-Playstore.csv") 
df.head()         
```

```python
df.tail(15)
```

**3- Checking the types of the data**


```python
df.dtypes
```

**4- Dropping irrelevant columns**

```python
pd.unique(df['Privacy Policy'])
```

```python
df = df.drop(['Privacy Policy'], axis=1)
df.head(5)
```

**5- Renaming the columns**

```python
df = df.rename(columns={"Rating Count":"rcount","Maximum Installs":"Max_installs" })
df.head(5)
```

**6- Dropping the duplicate rows**

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()    
```

```python
df = df.drop_duplicates()
df.head()
```

**7- Dropping the missing or null values**

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())  # After dropping the values
```

**8- Detecting Outliers**

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['Minimum Installs'])
```

**9- Plot different features against one another**

```python
df.Installs.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Install ratings ")
plt.ylabel('Ratings')
plt.xlabel('install');
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price'])
```

**9- Scatter Plot**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Max_installs'], y=df['rcount'])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['Rating'],y=df['Installs'])
```

**9- Heat Maps**

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
